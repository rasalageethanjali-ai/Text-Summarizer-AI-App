```python
from transformers import pipeline
import gradio as gr

# Load summarization pipeline (be patient, first-time model download may take 1â€“2 min)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize(text):
    summary = summarizer(text, max_length=60, min_length=20, do_sample=False)
    return summary[0]['summary_text']

demo = gr.Interface(
    fn=summarize,
    inputs="text",
    outputs="text",
    title="Text Summarizer (Hugging Face)",
    description="Enter English text and get a short summary powered by facebook/bart-large-cnn"
)
```

```python
demo.launch(share=True)
```
